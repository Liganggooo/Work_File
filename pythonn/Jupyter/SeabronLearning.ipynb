```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
tips = sns.load_dataset("tips")
```

```python
g=sns.FacetGrid(tips,col='day')
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(plt.hist, "tip");#加分号只会画出图像，不会储存，tip是自变量名，不是数据
```

```python
g = sns.FacetGrid(tips,col="sex",hue="smoker")
g.map(plt.scatter,'total_bill','tip',alpha=0.7)
g.add_legend();#增加右侧的标签
```

```python
g = sns.FacetGrid(tips,col='day',size=4,aspect=.5)
g.map(sns.barplot,'sex','total_bill');#sns.barplot是seaborn作直方图，plt.hist是matplotlib做直方图
```

```python
g = sns.FacetGrid(tips, row="smoker", col="time", margin_titles=True)
g.map(sns.regplot, "size", "total_bill", color=".3", fit_reg=False, x_jitter=.1);
```

```python
titanic = sns.load_dataset("titanic")
titanic = titanic.assign(deck=titanic.deck.astype(object)).sort_values("deck")
g = sns.FacetGrid(titanic, col="class", sharex=False,
                  gridspec_kws={"width_ratios": [5, 3, 3]})
g.map(sns.boxplot, "deck", "age");
```

```python
ordered_days = tips.day.value_counts().index
g = sns.FacetGrid(tips, row="day", row_order=ordered_days,
                  size=1.7, aspect=4,)
g.map(sns.distplot, "total_bill", hist=True, rug=True);#distplot带趋势线的直方图，hist默认为TRUE,是否带直方图，rug为图下刻度默认为false
```

```python
pal = dict(Lunch="seagreen", Dinner="gray")#选择标签的颜色
g = sns.FacetGrid(tips, hue="time", palette=pal, size=5)#用palette引入
g.map(plt.scatter, "total_bill", "tip", s=50, alpha=.7, linewidth=.5, edgecolor="white")
g.add_legend();
```

```python
g = sns.FacetGrid(tips, hue="sex", palette="Set1", size=5, hue_kws={"marker": ["^", "v"]})#hue为色调，表示第三个维度变量
g.map(plt.scatter, "total_bill", "tip", s=100, linewidth=.5, edgecolor="white")#s控制图中点的大小
g.add_legend();
```

```python
attend = sns.load_dataset("attention").query("subject <= 12")
g = sns.FacetGrid(attend, col="subject", col_wrap=4, size=2, ylim=(0, 10))
g.map(sns.pointplot, "solutions", "score", color=".3", ci=None);
```

```python
with sns.axes_style("white"):
    g = sns.FacetGrid(tips, row="sex", col="smoker", margin_titles=True, size=2.5)
g.map(plt.scatter, "total_bill", "tip", color="#334488", edgecolor="white", lw=.5);
g.set_axis_labels("Total bill (US Dollars)", "Tip");
g.set(xticks=[10, 30, 50], yticks=[2, 6, 10]);
g.fig.subplots_adjust(wspace=.02, hspace=.02);
```
