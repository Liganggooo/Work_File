```python
isCN  = 1
font_path = 'font.ttf'
stopwords_path = 'stopwords.txt'
Imgname1 = 'GK1.png'
Imgname2 = 'GK2.png'
background_img = 'GK3.png'
userdict = 'userdict.txt'
word_file = 'txt_freq.txt'
```

```python
text = open(word_file,'r',encoding = 'utf-8').read()
text = text.replace('\n','')
text = text.replace('\ufeff','')
# text
```

```python
# alice_coloring = np.array(Image.open(background_img))
# stopwords = set(STOPWORDS)
# stopwords.add('said')
wc = WordCloud( font_path = font_path,
               background_color = 'white',
               width = 2000,height =1000,
               max_words=2000,
               max_font_size=400,
               collocations=False,#去重
               random_state = 30,
              )
# wc.generate(text)
jieba.load_userdict(userdict)
```

```python
def jiebacleartext():
    mywordlist = []
    seg_list = jieba.cut(text, cut_all=False)
    liststr="/ ".join(seg_list)
    f_stop = open(stopwords_path,'r',encoding = 'utf-8')
    try:
        f_stop_text = f_stop.read()
#         f_stop_text = unicode(f_stop_text,'utf-8')
    finally:
        f_stop.close()
    f_stop_seg_list = f_stop_text.split('\n')
    for myword in liststr.split('/'):
        if not(myword.strip() in f_stop_seg_list) and len(myword.strip()) > 1:
            mywordlist.append(myword)
    return ''.join(mywordlist)
```

```python
if isCN:
    baobao = jiebacleartext()
    print(baobao)
```

```python
# wc.generate_from_frequencies(text)
wc.generate(baobao)
plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()
wc.to_file(Imgname2)
```
