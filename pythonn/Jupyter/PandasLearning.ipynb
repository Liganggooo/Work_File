```python
zips = content['Incident Zip']
# 用is_close表示0或者1开始的比较正确的邮编
is_close = zips.str.startswith('0') | zips.str.startswith('1')
# 非缺省值但不以0或者1开始的邮编认为是有些困惑的
is_far = ~(is_close) & zips.notnull()
```

```python
na_values = ['NO CLUE', 'N/A', '0']
content = pd.read_csv(r'C:\Users\lg\Desktop\Pythonn\Python\311-service-requests.csv',
                      sep=',',na_values = na_values, dtype={'Incident Zip': str})
def fix_zip_codes(zips):
    # Truncate everything to length 5 
    zips = zips.str.slice(0, 5)
    
    # Set 00000 zip codes to nan
    zero_zips = zips == '00000'
    zips[zero_zips] = np.nan
    
    return zips
```

```python
content['Incident Zip'] = fix_zip_codes(content['Incident Zip'])
```
