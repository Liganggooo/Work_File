```python
%matplotlib inline
import os
import pandas as pd
from pandas import Series,DataFrame
# os.chdir('desktop\dailydoc\管委会调查')
os.getcwd()
```

```python
xls_file = pd.ExcelFile('name.xlsx')
table = xls_file.parse('Sheet1')
table.head()
```

```python
table1 = table[['所在省','省合计']]
deal_data = table1.dropna()#删除空行
# deal_data.order()#对序列值进行排序
data = deal_data.sort_index(by = '省合计',ascending = False)
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# plt.rcParams['font.sans-serif'] = ['SimHei']  # 中文字体设置-黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
sns.set(font='SimHei')  # 解决Seaborn中文显示问题
g = sns.FacetGrid(data,size=10,aspect=2)
g.map(sns.barplot,'所在省','省合计');#sns.barplot是seaborn作直方图，plt.hist是matplotlib做直方图
```
