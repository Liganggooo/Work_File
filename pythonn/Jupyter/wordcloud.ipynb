```python
%matplotlib inline
import os
import jieba
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
import jieba.analyse as analyse
```

```python
isCN  = 1
font_path = 'font.ttf'
stopwords_path = 'stopwords.txt'
Imgname1 = 'WordCloudDefautColors.png'
Imgname2 = 'WordCloudColorsByImg.png'
background_img = 'benren.png'
userdict = 'userdict.txt'
word_file = 'magic.txt'
```

```python
text = open(word_file,'r',encoding = 'utf-8').read()
text = text.replace('\n\u3000\u3000','')
text = text.replace('\ufeff','')
```

```python
alice_coloring = np.array(Image.open(background_img))
# stopwords = set(STOPWORDS)
# stopwords.add('said')
wc = WordCloud(background_color = 'white',
               font_path = font_path,
               width = 2000,height =1000,
               max_words=2000,
               mask = alice_coloring,
               max_font_size=400, 
               random_state=42)
# wc.generate(text)
image_color = ImageColorGenerator(alice_coloring)
jieba.load_userdict(userdict)
```

```python
jieba.load_userdict(userdict)
wordlist = jieba.cut(text,cut_all = False,HMM = True)
magic = ' '.join(wordlist)
# for key in analyse.extract_tags(text,50,withWeight = False):
#     print(key)
```

```python
def jiebacleartext():
    mywordlist = []
    seg_list = jieba.cut(text, cut_all=False)
    liststr="/ ".join(seg_list)
    f_stop = open(stopwords_path,'r',encoding = 'utf-8')
    try:
        f_stop_text = f_stop.read()
#         f_stop_text = unicode(f_stop_text,'utf-8')
    finally:
        f_stop.close()
    f_stop_seg_list = f_stop_text.split('\n')
    for myword in liststr.split('/'):
        if not(myword.strip() in f_stop_seg_list) and len(myword.strip()) > 1:
            mywordlist.append(myword)
    return ''.join(mywordlist)
```

```python
if isCN:
    threeBody = jiebacleartext()
#     print(threeBody)
```

```python
wc.generate(threeBody)
plt.imshow(wc)
plt.axis('off')
plt.figure
wc.to_file(Imgname1)
```
